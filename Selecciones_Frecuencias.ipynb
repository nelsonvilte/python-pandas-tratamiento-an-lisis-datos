# Selecciones y Frecuencias

```python
import pandas as pd
```

```python
datos =pd.read_csv('alquiler_residencial.csv',sep=';')
datos.head(10)
```

```python
#Selecione solamente los imuebles clasificados con tipo 'Departamento'
seleccion = datos['Tipo']=='Departamento'
seleccion #los que cumplen el criterio son True

```

```python
n1 = datos[seleccion].shape[0] #los inmuebles tipo departamento, la frecuencia
n1
```

```python
#Selecione solamente los imuebles clasificados con tipo 'Casa', 'Casa de condominio' y 'Casa de villa'
seleccion = (datos['Tipo']=='Casa')|(datos['Tipo']=='Casa en condominio')|(datos['Tipo']=='Casa de villa')
seleccion #los que cumplen el criterio son True
```

```python
datos[seleccion]
```

```python
n2 = datos[seleccion].shape[0]
n2
```

```python
#Seleccione los inmuebles con área entre 60 y 100  metros cuadrados, incluyendo los limites
seleccion = (datos['Area']>=60) & (datos['Area']<=100)
n3 = datos[seleccion].shape[0]
n3
```

```python
#Seleccione los inmuebles que tengan por lo menos 4 cuartos y alquiler menor que $2.000,00

seleccion = (datos['Cuartos']>=4) & (datos['Valor']<2000)
n4 = datos[seleccion].shape[0]
n4
```

```python
print("N° de inmuebles clasificados con tipo 'Departamento' -> {}".format(n1))
print("N° de inmuebles clasificados con tipo 'Casa', 'Casa de condominio' y 'Casa de villa' -> {}".format(n2))
print("N° de inmuebles con área entre 60 y 100  metros cuadrados, incluyendo los limites -> {}".format(n3))
print("N° de inmuebles que tengan por lo menos 4 cuartos y alquiler menor que $2.000,00 -> {}".format(n4))
```

# Ejercicios con Dataframes  utilizando técnicas de selección

```python
#Dataframe
import pandas as pd
alumnos = pd.DataFrame({
'Nombre': ['Ary', 'Katia', 'Denis', 'Beto', 'Bruna', 'Dara', 'Carlos', 'Alice'], 
'Sexo': ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F'], 
'Edad': [15, 27, 56, 32, 42, 21, 19, 35], 
'Notas': [7.5, 2.5, 5.0, 10, 8.2, 7, 6, 5.6], 
'Aprobado': [True, False, False, True, True, True, False, False]},
columns = ['Nombre', 'Edad', 'Sexo', 'Notas', 'Aprobado'])
```

```python
alumnos
```

**Selección Nivel 1**

- Estudiantes aprobados

```python


seleccion = alumnos['Aprobado'] == True
aprobados = alumnos[seleccion]
aprobados
```

**Selección Nivel 2**

- DataFrame sólo con las estudiantes aprobadas

```python


seleccion = (alumnos.Aprobado == True) & (alumnos.Sexo == 'F')
aprobadas = alumnos[seleccion]
aprobadas
```

**Selección Nivel 3**

- Solo una visualización de los alumnos con edades entre 10 a 20 años o con edad mayor/igual a 40 años.

```python
seleccion = (alumnos.Edad > 10) & (alumnos.Edad < 20) | (alumnos.Edad >= 40)
alumnos[seleccion]
```

**Seleccion Nivel 4**

- DataFrame sólo con los estudiantes reprobados y mantenga en este DataFrame sólo las columnas Nombre, Sexo y Edad, en este orden.



```python
#Alternativa 1
seleccion = alumnos['Aprobado'] == False
reprobados = alumnos[['Nombre', 'Sexo', 'Edad']][seleccion]
reprobados
```

```python
#Alternativa 2
seleccion = alumnos['Aprobado'] == False
reprobados = alumnos.loc[seleccion, ['Nombre', 'Sexo', 'Edad']]
reprobados
```

**Seleccion Nivel 5**

- Visualización con los tres alumnos más jóvenes.


```python
alumnos.sort_values(by = ['Edad'], inplace = True)
alumnos.iloc[:3]
```
