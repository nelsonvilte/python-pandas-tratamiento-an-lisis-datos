**Determinando selecciones**

Crea un DataFrame a partir de letras y números.


```python
import pandas as pd

numeros = [i for i in range(11)]
letras = [chr(i + 65) for i in range(11)]
nome_coluna = ['N']

df = pd.DataFrame(data = numeros, index = letras, columns = nome_coluna)
```

Basándonos en este Dataframe, generamos una selección para filtrar los datos:

```python
seleccion= df['N'].isin([i for i in range(11) if i % 2 == 0])
df = df[seleccion]
df
```

El código anterior crea un DataFrame con una columna llamada N y le asigna los números del 0 al 10. Con el método isin(), seleccionamos sólo los números pares de N.

# Organizando Dataframes (Sort)

```python
import pandas as pd
```

```python
data = [[1,2,3],[4,5,6],[7,8,9]]
data
```

```python
list('321')
```

```python
df = pd.DataFrame(data, list('321'), list('ZYX'))
df
```

```python
df.sort_index() 
df
```

```python
df.sort_index(inplace=True) #ordenar el dataframe por el nombre de las filas
df
```

```python
df.sort_index(inplace=True, axis = 1) #ordenar el dataframe por el nombre de las columnas
df
```

```python
df.sort_values(by = 'X',inplace=True) #ordenar el dataframe por el contenido, verticalmente
df
```

```python
df.sort_values(by = '3',axis =1,inplace=True) #ordenar el dataframe por el contenido, horizontalmete
df
```
