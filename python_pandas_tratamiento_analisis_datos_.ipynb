# Crear Dataframes de diferentes tipos de archivos

```python
import pandas as pd
```

```python
json = open('datos/alquiler.json')
print(json.read())
```

```python
df_json = pd.read_json('datos/alquiler.json')
df_json
```

```python
txt = open('datos/alquiler.txt')
print(txt.read())
```

```python
df_txt = pd.read_table('datos/alquiler.txt')
df_txt
```

```python
df_xlsx= pd.read_excel('datos/alquiler.xlsx')
df_xlsx
```

```python
df_html= pd.read_html('datos/datos_html_1.html')
df_html[0]
```


*   Cuando usamos la funcion read_html() en una url o archivo HTML que contenga mas de una tabla, cada una es convertida en un DataFrame y todas se almacenan en una lista de Python, es decir, la funcion read_html(), devuelve una lista de DataFrames.


*   Cuando se usa la funcion read_html(), es posible pasar como parametro de entrad(fuente de los datos) una url.
